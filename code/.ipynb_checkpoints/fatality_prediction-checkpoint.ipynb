```python
import pandas as pd
import numpy as np
```

```python
accidents = pd.read_csv('../data/accidents.csv')
casualties = pd.read_csv('../data/casualties.csv', on_bad_lines = 'skip')
vehicles = pd.read_csv('../data/vehicles.csv', on_bad_lines = 'skip')
```

```python
main_df = accidents.merge(casualties).merge(vehicles)
```

## Remove columns with lots or few unique values

```python
for i in range(main_df.shape[1]):
    unique_vals = len(np.unique(main_df[main_df.columns[i]].astype('str')))
    percentage = unique_vals/main_df.shape[0]
    print(f'{main_df.columns[i]}: {unique_vals}, {percentage*100}%')
```
